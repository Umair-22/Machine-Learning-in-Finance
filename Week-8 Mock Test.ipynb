```python
# Load the data
df_2023 = pd.read_csv('df_2023.csv')

# Convert 'Local time' to datetime if needed
df_2023['Local time'] = pd.to_datetime(df_2023['Local time'])

# Plot price data
plt.figure(figsize=(14, 8))

# Subplot 1: Bid and Ask Prices
plt.subplot(2, 1, 1)
plt.plot(df_2023['Local time'], df_2023['Open_Bid'], label='Open Bid', color='blue')
plt.plot(df_2023['Local time'], df_2023['High_Bid'], label='High Bid', color='cyan')
plt.plot(df_2023['Local time'], df_2023['Low_Bid'], label='Low Bid', color='green')
plt.plot(df_2023['Local time'], df_2023['Close_Bid'], label='Close Bid', color='purple')

plt.plot(df_2023['Local time'], df_2023['Open_Ask'], label='Open Ask', color='orange')
plt.plot(df_2023['Local time'], df_2023['High_Ask'], label='High Ask', color='red')
plt.plot(df_2023['Local time'], df_2023['Low_Ask'], label='Low Ask', color='brown')
plt.plot(df_2023['Local time'], df_2023['Close_Ask'], label='Close Ask', color='pink')

plt.title('Price Chart')
plt.xlabel('Local Time')
plt.ylabel('Price')
plt.legend()
plt.grid()

# Subplot 2: Volume Data
plt.subplot(2, 1, 2)
plt.plot(df_2023['Local time'], df_2023['Volume_Bid'], label='Volume Bid', color='blue')
plt.plot(df_2023['Local time'], df_2023['Volume_Ask'], label='Volume Ask', color='orange')

plt.title('Volume Chart')
plt.xlabel('Local Time')
plt.ylabel('Volume')
plt.legend()
plt.grid()

# Display the plot
plt.tight_layout()
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler

# Select columns to normalize (all except 'Local time')
columns_to_normalize = ['Open_Bid', 'High_Bid', 'Low_Bid', 'Close_Bid', 'Volume_Bid', 
                        'Open_Ask', 'High_Ask', 'Low_Ask', 'Close_Ask', 'Volume_Ask']

# Initialize the MinMaxScaler
scaler = MinMaxScaler()

# Fit and transform the data
df_2023[columns_to_normalize] = scaler.fit_transform(df_2023[columns_to_normalize])

# Display the normalized data
print(df_2023.head())
```

```python
# Define the columns containing prices
price_columns = ['Open_Bid', 'High_Bid', 'Low_Bid', 'Close_Bid', 
                 'Open_Ask', 'High_Ask', 'Low_Ask', 'Close_Ask']

# Define the shift amount
shift_amount = 10

# Shift the minimum and maximum prices in each price column
for col in price_columns:
    # Calculate the minimum and maximum values
    min_value = df_2023[col].min()
    max_value = df_2023[col].max()
    
    # Shift the values in the column
    df_2023[col] = df_2023[col].apply(lambda x: x - shift_amount if x == min_value else (x + shift_amount if x == max_value else x))

# Display the modified data
print(df_2023.head())
```

```python
# Define input and output columns
input_columns = ['Open_Bid', 'High_Bid', 'Low_Bid', 'Volume_Bid', 
                 'Open_Ask', 'High_Ask', 'Low_Ask', 'Volume_Ask']
output_columns = ['Close_Bid', 'Close_Ask']

# Separate inputs and outputs
X = df_2023[input_columns]  # Input matrix
y = df_2023[output_columns]  # Output matrix

# Display the first few rows of inputs and outputs
print("Input matrix (X):")
print(X.head())
print("\nOutput matrix (y):")
print(y.head())
```

```python
data = df_2023[input_columns].values  # Convert to NumPy array

# Define the number of timesteps for each sample in the 3D tensor
timesteps = 5

# Create a 3D tensor using sliding windows
samples = data.shape[0] - timesteps + 1  # Number of samples in the tensor
features = data.shape[1]  # Number of features

# Initialize the tensor with zeros
tensor_3d = np.zeros((samples, timesteps, features))

# Populate the 3D tensor using a sliding window approach
for i in range(samples):
    tensor_3d[i] = data[i:i + timesteps]

# Display the shape of the resulting tensor and first sample
print("Shape of 3D tensor:", tensor_3d.shape)
print("First sample in 3D tensor:\n", tensor_3d[0])
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

# Assuming `tensor_3d` is your input tensor and `y` is the output matrix

# Define the model
model = Sequential([
    LSTM(64, input_shape=(tensor_3d.shape[1], tensor_3d.shape[2]), return_sequences=False),
    Dense(32, activation='relu'),
    Dense(y.shape[1])  # Output layer, matching the number of output features
])

# Compile the model
model.compile(optimizer='adam', loss='mse')

# Train the model
model.fit(tensor_3d, y[timesteps - 1:], epochs=10, batch_size=16, validation_split=0.2)

# Display model summary
model.summary()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(tensor_3d, y[timesteps - 1:], test_size=0.2, random_state=42)

# Train the model on the training set
model.fit(X_train, y_train, epochs=10, batch_size=16, validation_split=0.2)

# Make predictions on the test set
y_pred = model.predict(X_test)

# Calculate MSE and MAE
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)

print("Mean Squared Error (MSE):", mse)
print("Mean Absolute Error (MAE):", mae)
```

```python
import matplotlib.pyplot as plt

# Convert y_test and y_pred to NumPy arrays, if they arenâ€™t already
y_test = y_test.to_numpy() if isinstance(y_test, pd.DataFrame) else y_test
y_pred = y_pred.to_numpy() if isinstance(y_pred, pd.DataFrame) else y_pred

# Plot the actual vs. predicted values for each output variable
plt.figure(figsize=(14, 6))

# Loop through each output variable to plot them separately
for i in range(y_test.shape[1]):
    plt.subplot(1, y_test.shape[1], i + 1)
    plt.plot(y_test[:, i], label='Actual', color='blue')
    plt.plot(y_pred[:, i], label='Predicted', color='orange')
    plt.title(f'Output Variable {i + 1}')
    plt.xlabel('Sample')
    plt.ylabel('Value')
    plt.legend()

# Display the plot
plt.tight_layout()
plt.show()
```
